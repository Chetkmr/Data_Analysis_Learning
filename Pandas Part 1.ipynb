```python
# Groupby
data=pd.read_excel("E:\Data Analyst\Resources\Data Sets\Data set of Data Analytics FULL Course for Beginners to Pro in 29 HOURS 2025 Edition\Datasets-main\ESD.xlsx")
# print(data.head(10))

gp=data.groupby(["Department", "Gender"]).agg({"EEID":"count"})
# print(gp)
print("~~~~~~~~~~")

gp=data.groupby(["Country", "Gender","Department"]).agg({"Age":"count", "Age":"max"})
gp=gp[(gp["Age"]>50)|(gp["Age"]<40)]
print(gp)
```
