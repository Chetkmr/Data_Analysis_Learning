````````python
import numpy as np

a=[30,40,20],[30,30,50]
arr=np.array(a)
print(arr)
print(arr.shape) #rows, coloums
print(len(arr)) #number or nested values
print(np.size(arr)) #number of size
print(type(arr)) #type of variable
print(arr.dtype) #datatype of array
print(arr.astype(float)) #converction of atatype
print(arr.astype(str))
````````

````````python
##math and operatio with function 

arr1 = np.array([[30,30,40],[40,20,29]])
arr2 = np.array([[40,50,20],[30,50,10]])
print(arr1+arr2)
print(np.add(arr1, arr2))
print("```````")

print(arr1-arr2)
print(np.subtract(arr1, arr2))
print("```````")

print(arr1*arr2)
print(np.multiply(arr1, arr2))
print("```````")

print(arr1/arr2)
print(np.divide(arr1, arr2))
print("```````")

arr3 = np.array([3])
print(np.power(arr1,arr3))
print("```````")

arr4 = np.array([64000, 8000, 24389])
print(np.sqrt(arr4).astype(int))
#print(astype())
````````

````````python
##combining and spliting array

## concatenate 

arr1 = np.array([2,5,4,1,6,3,2])
arr2 = np.array([2,5,4,1,6,3,2])
print(np.concatenate([arr1,arr2]))
print("```````")

#axis 1 Horizontal 2 vertical
arr3 = np.array([[2,5,4],[1,6,3]])
arr4 = np.array([[2,5,4],[1,6,3]])

print(np.concatenate([arr3,arr4]))
print("```````")
print(np.concatenate([arr3, arr4], axis =0))
print("```````")
print(np.concatenate([arr3, arr4], axis =1))
print("```````")

print(np.hstack([arr3,arr4])) ##horizontal concatenation
print("```````")

print(np.vstack([arr3,arr4])) ##vertical concatenation
print("```````")

## spliting

arr5 =np.array([5,4,6,2,4,8])
print(np.array_split(arr5,3))
print("```````")
## or
arr5a = np.array_split(arr5, 3) 
print("index 0",arr5a[0]) ##index
print("index 1",arr5a[1]) ##index
print("index 2",arr5a[2]) ##index
print("```````")
````````
